```python
#The client wants a house either with a very beautiful view from the window or with a view of the waterfront, 
#in very good condition and built at least 1980. In what price range will the houses be?
```

```python
kc_house[((kc_house['view']==4) | (kc_house['waterfront']==1)) & (kc_house['condition']==5) & (kc_house['yr_built']>1980)].price
```

```python
#The client wants a house without a basement, with two floors, costing up to 150,000. What is the average condition rating of such houses?
```

```python
kc_house[(kc_house['sqft_basement']==0) & (kc_house['floors']==2) & (kc_house['price']<150000)].condition.mean()
```
